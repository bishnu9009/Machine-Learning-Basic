<h6> Q1. Find and treat all the missing values. Rows or columns with missing values must not be dropped.<br></h6>


```python
#Answer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Dataset_Day4.csv') # Creating data frame from the csv.

df.info()
```

```python
print(df.head(50))                  # Data Frame with the NaN values.

print(f'\nThe DataFrame has the following number of NaN values in each column:-\n{df.isna().sum()}')
```

```python
df.fillna(df.mean(),inplace=True)    # Using imputation for the missing values.
                                     # It is a process where missing values are replaced with estimated or
                                     # calculated values based on the available data
```

```python
print(df.head(50))                   # Data Frame with-out the na values.

print(f'\nThe DataFrame has the following number of NaN values in each column:-\n{df.isna().sum()}')
```

<b>
    It can be seen that the NaN values are removed from the dataframe and it is replaced by the mean of each column.
