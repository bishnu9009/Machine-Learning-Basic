```python
#Answer
print("\n\nBelow are the boxplots of each variables..\n")   # Vizualisation as boxplot of all the variables
for column in df.columns:
    plt.figure()
    plt.boxplot(df[column])
    plt.title("Box Plot of " + column)
    plt.xlabel(column)
plt.show()
```

<h6>Question 2:- Create a simple linear regression model that quantitatively relates â€˜MEDVâ€™ with â€˜RMâ€™. (10 marks)<br><br>
a.	Share the model performance metrics and print the full regression model with coefficients.<br>
b.	Use the model to predict the price of the house for â€˜RMâ€™ = 7</h6>

```python
#Outliers

df["RM"] = (df["RM"]-df["RM"].mean())/df["RM"].std()


OutlierRows = df[(df.RM < -3) | (df.RM > 3)]

df_NoOutlier = df.drop(OutlierRows.index,axis=0)
df_NoOutlier.head()
df_NoOutlier.info()
```

498 Entries present after removal of outliers so 506-498=8 outliers

```python
# Data Splitting
from sklearn.model_selection import train_test_split

x = pd.DataFrame(df_NoOutlier["RM"])     # Making dataframe x from the RM columns of NoOutLiers dataframe.

y = pd.DataFrame(df_NoOutlier["MEDV"])   # Making dataframe y from the MEDV columns of NoOutLiers dataframe.

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)

x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error, mean_squared_error

lm = LinearRegression()                            # Calling and fitting the model on training dataset.
lm = lm.fit(x_train,y_train)

y_pred = lm.predict(x_test)                        # Predicting using test dataset

print(f"Scale Parameter:- {lm.coef_}\n")           # Scale parameter
print(f"Intercept Parameter:- {lm.intercept_}\n")  # Intercept parameter


print(f"R^2 Score:- {r2_score(y_test, y_pred)}\n")

print(f"The Mean Absolute Error:- {mean_absolute_error(y_test, y_pred)}\n")
print(f"The Mean Squared Error:- {mean_squared_error(y_test, y_pred)}\n")

rm_new = [[7]]                                     # Use the model to predict the price for RM = 7
price_pred = model.predict(rm_new)
print("Predicted Price for 'RM = 7':-$", price_pred[0]*1000)
```
